# Notebook 2: Baseline GNN Detector

Train and evaluate the baseline GNN-based anomaly detector.

```python
import sys, os
sys.path.insert(0, os.path.dirname(os.getcwd()))

import torch
import matplotlib.pyplot as plt

from src.dataset.loader import get_dataset
from src.models.detector import GCNDetector, GATDetector
from src.train import train_model
from src.utils.visualization import plot_training_curve
```

```python
# Load data
data, labels, desc = get_dataset(seed=42)
print(desc)

# Device
device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(f"Device: {device}")
```

```python
# Initialize model
model = GCNDetector(
    in_channels=16,
    hidden_channels=64,
    out_channels=3,
    dropout=0.5,
)
print(f"Parameters: {sum(p.numel() for p in model.parameters()):,}")
```

```python
# Train
model, history, metrics = train_model(
    model, data,
    n_epochs=200,
    lr=0.01,
    device=device,
)
```

```python
# Plot training curve
plot_training_curve(
    history["train_loss"],
    history["val_loss"],
    history["val_acc"],
    save_path="training_curve.png",
)
```

```python
# Save model
torch.save({
    "model_state": model.state_dict(),
    "metrics": metrics,
}, "baseline_model.pth")
print("Model saved to baseline_model.pth")
```
